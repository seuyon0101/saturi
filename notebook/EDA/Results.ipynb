```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
```

```python
path = os.getenv('HOME') + '/aiffel/DATA'
```

```python
files = glob.glob(path + '/*results*.csv')
```

```python
results = ['cmsp8000','spm8000','spm4000','cmsp8000max','cmsp4000','cmsp4000max','msp4000']
```

```python
df = pd.read_csv(files[0],usecols=['Unnamed: 0','reg','tok_cat','topic','eng','dial','bleu','pred'], index_col = 'Unnamed: 0')
df.rename(columns={'bleu':'bleu_cmsp8000','pred' : 'pred_cmsp8000'}, inplace=True)

for file, res in zip(files[1:], results[1:]) :
    temp_ = pd.read_csv(file, usecols=['Unnamed: 0','reg','tok_cat','topic','eng','dial','bleu','pred'], index_col = 'Unnamed: 0')
    temp_.rename(columns={'bleu':f'bleu_{res}','pred' : f'pred_{res}'},inplace=True)
    df = pd.merge(df, temp_.iloc[:,[-1,-2]], right_index=True, left_index=True)
reorder_col = ['reg','tok_cat','topic','eng','dial','pred_spm4000','pred_msp4000','pred_cmsp4000','pred_cmsp4000max','pred_spm8000','pred_cmsp8000','pred_cmsp8000max','bleu_spm4000','bleu_msp4000','bleu_cmsp4000','bleu_cmsp4000max','bleu_spm8000','bleu_cmsp8000','bleu_cmsp8000max']
df = df[reorder_col].copy()
```
